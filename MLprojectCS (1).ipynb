```python
CustomerData=pd.merge(CustomerData,cancelledTransactions,on='CustomerID',how='left')
```

```python
finalData=CustomerData.drop('CustomerID',axis=1)
finalData
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
km=KMeans(init='k-means++',n_init=6,max_iter=100,random_state=0)
K_Visualizer=KElbowVisualizer(km,k=(2,15),timings=False)
```

```python
import matplotlib.pyplot as plt
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(finalData))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

```python
K_Visualizer.fit(finalData)
```

```python
model=KMeans(n_clusters=5)
```

```python
model.fit(finalData)
```

```python
centroids=model.cluster_centers_
labels=model.labels_
```
